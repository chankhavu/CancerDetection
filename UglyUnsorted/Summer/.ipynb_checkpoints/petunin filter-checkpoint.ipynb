```python
import numpy as np
from matplotlib import image as mpimg
from matplotlib import pyplot as plt
import cv2
import math
from scipy.spatial import ConvexHull

%pylab inline
```

```python
img = mpimg.imread('1_1.bmp')
imgplot = imshow(img)
```

```python
def kernel_filter(img, ordering, size):
    new_img = numpy.ndarray(img.shape, dtype=uint8)    
    def apply_kernel(img, new_img, ordering, x, y, size):     
        assert size % 2 == 1
        top = max(y-(size//2), 0)
        bottom = min(y+(size//2), img.shape[0])
        left = max(x-(size//2), 0)
        right = min(x+(size//2), img.shape[1])        
        
        pixels = []        
        for i in range(top, bottom):
            for j in range(left, right):
                pixels += [img[i][j]]
        value = ordering(pixels)
                
        pixel_value = value #pixels[len(pixels)//2 + 1]
        new_img[y][x] = pixel_value

    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            apply_kernel(img, new_img, ordering, j, i, size)
    return new_img
```

```python
def normal_ordering(pixels):
    ordered = sorted(pixels, key=lambda point: 0.33*point[0] + 0.33*point[1] + 0.34*point[2])
    return ordered[len(ordered)//2 + 1]

pylab.rcParams['figure.figsize'] = (6, 6)
filtered_img = kernel_filter(img, normal_ordering, 7)
imgplot = imshow(filtered_img)
plt.title('Normal ordering')
```

```python
def marginal_ordering(pixels):    
    p_red = [p[0] for p in pixels]
    p_green = [p[1] for p in pixels]
    p_blue = [p[2] for p in pixels]
    
    if len(pixels) % 2 == 1:
        middle_red = p_red[len(p_red)//2 + 1]
        middle_green = p_green[len(p_green)//2 + 1]        
        middle_blue = p_blue[len(p_blue)//2 + 1]
    else:
        middle_red = (int(p_red[len(p_red)//2]) + p_red[len(p_red)//2 + 1]) // 2
        middle_green = (int(p_green[len(p_green)//2]) + p_green[len(p_green)//2 + 1]) // 2
        middle_blue = (int(p_blue[len(p_blue)//2]) + p_blue[len(p_blue)//2 + 1]) // 2
            
    avg_red = sum(p_red) // len(p_red)
    avg_green = sum(p_green) // len(p_green)
    avg_blue = sum(p_blue) // len(p_blue)
    
    p_red.sort()
    p_green.sort()
    p_blue.sort()
    if len(pixels) % 2 == 1:
        median_red = p_red[len(p_red)//2 + 1]
        median_green = p_green[len(p_green)//2 + 1]
        median_blue = p_blue[len(p_blue)//2 + 1]
    else:
        median_red = (int(p_red[len(p_red)//2 + 1]) + p_red[len(p_red)//2]) // 2
        median_green = (int(p_green[len(p_green)//2 + 1]) + p_green[len(p_green)//2]) // 2
        median_blue = (int(p_blue[len(p_blue)//2 + 1]) + p_blue[len(p_blue)//2]) // 2
        
    pr = min(p_red, key=lambda pixel: abs(int(pixel) - middle_red) + abs(int(pixel) - avg_red) + abs(int(pixel) - median_red))
    pg = min(p_green, key=lambda pixel: abs(int(pixel) - middle_green) + abs(int(pixel) - avg_green) + abs(int(pixel) - median_green))
    pb = min(p_blue, key=lambda pixel: abs(int(pixel) - middle_blue) + abs(int(pixel) - avg_blue) + abs(int(pixel) - median_blue))
    
    answer = [pr, pg, pb]  #[[p_red[i], p_green[i], p_blue[i]] for i in range(len(pixels))]
    return answer
    
pylab.rcParams['figure.figsize'] = (6, 6)
filtered_img = kernel_filter(img, marginal_ordering, 7)
imgplot = imshow(filtered_img)
plt.title('Marginal ordering')
```

```python
def move(points, vec):
    assert vec.shape[0] == points.shape[1]    
    for p in points:
        p = p + vec
        
def rotate(points, xy_angle, xz_angle, center = np.array([0, 0, 0], dtype=np.float_)):
    # rotate in XY plane
    for p in points:
        p_ = p = (p - center)        
        p_[0] = p[0] * math.cos(xy_angle) - p[1] * math.sin(xy_angle)
        p_[1] = p[0] * math.sin(xy_angle) + p[1] * math.cos(xy_angle)
        p = (center + p_)
        
    # rotate in XZ plane
    for p in points:
        p_ = p = (p - center)        
        p_[0] = p[0] * math.cos(xy_angle) - p[2] * math.sin(xy_angle)
        p_[2] = p[0] * math.sin(xy_angle) + p[2] * math.cos(xy_angle)
        p = (center + p_)
    

def petunin_ordering(int_pixels):
    # hull = ConvexHull(pixels)
    
    pixels = np.array(int_pixels, dtype=np.float_)    
    
    # O(n^2) algorithm
    max_squared_distance = 0
    dp0, dp1 = None, None
    for p0 in pixels:
        for p1 in pixels:
            d_squared = (p0[0] - p1[0])**2 + (p0[1] - p1[1])**2 + (p0[2] - p1[2])**2
            if d_squared > max_squared_distance: 
                max_squared_distance = d_squared
                dp0, dp1 = p0, p1
                
    # move the diameter to [0, 0] point
    vec = -dp0
    move(pixels, vec)
    p0, p1 = np.array([0, 0, 0]), dp1 + vec
    
    """
    # rotate so that the diameter lies on X axis
    v1xy = np.array([p1[0] - p0[0], p1[1] - p0[1]])
    v0xy = np.array([1, 0])
    if (v1xy[0]*v0xy[0] + v1xy[1]*v0xy[1] == 0) and (
            math.hypot(v1xy[0], v1xy[1]) == 0 or math.hypot(v0xy[0], v0xy[1]) == 0):
        xy_angle = 0
    else:
        xy_angle = -math.acos( float(v1xy[0]*v0xy[0] + v1xy[1]*v0xy[1])/
                        (math.hypot(v1xy[0], v1xy[1])*math.hypot(v0xy[0], v0xy[1])) )        
    v1xz = np.array([p1[0] - p0[0], p1[2] - p0[2]])
    v0xz = np.array([1, 0])
    if (v1xz[0]*v0xz[0] + v1xz[1]*v0xz[1] == 0) and (
            math.hypot(v1xz[0], v1xz[1]) == 0 or math.hypot(v0xz[0], v0xz[1]) == 0):
        xz_angle = 0
    else:
        xz_angle = -math.acos( float(v1xz[0]*v0xz[0] + v1xz[1]*v0xz[1])/
                        (math.hypot(v1xz[0], v1xz[1])*math.hypot(v0xz[0], v0xz[1])) )    
    """
    
    v1xy = np.array([p1[0] - p0[0], p1[1] - p0[1]])
    v0xy = np.array([1, 0])
    dot_xy = v0xy[0]*v1xy[0] + v0xy[1]*v1xy[1]
    det_xy = v0xy[0]*v1xy[1] + v0xy[1]*v1xy[0]
    xy_angle = math.atan2(det_xy, dot_xy)
    
    v1xz = np.array([p1[0] - p0[0], p1[2] - p0[2]])
    v0xz = np.array([1, 0])
    dot_xz = v0xz[0]*v1xz[0] + v0xz[1]*v1xz[1]
    det_xz = v0xz[0]*v1xz[1] + v0xz[1]*v1xz[0]
    xz_angle = math.atan2(det_xz, dot_xz)
    
    d = np.array([p1[0] - p0[0], p1[1] - p0[1], p1[2] - p0[2]])
    xy_angle = -math.atan2(d[1], d[0])
    dx = d[0] * math.cos(xy_angle) - d[1] * math.sin(xy_angle)
    dy = d[0] * math.sin(xy_angle) + d[1] * math.cos(xy_angle)
    d[0], d[1] = dx, dy
    xz_angle = -math.atan2(d[2], d[0])
    rotate(pixels, xy_angle, xz_angle)
        
    assert(abs(d[1])< 0.01)
    
    # get boundary rectangle
    max_x, min_x = max([p[0] for p in pixels]), min([p[0] for p in pixels])
    max_y, min_y = max([p[1] for p in pixels]), min([p[1] for p in pixels])
    max_z, min_z = max([p[2] for p in pixels]), min([p[2] for p in pixels])
    
    xm, ym, zm = (max_x + min_x / 2), (max_y + min_y / 2), (max_z + min_z / 2)
    pdist = np.array([( ((p[0]-xm)/(1))**2 + ((p[1]-ym)/(1))**2 + ((p[2]-zm)/(1))**2 ) for p in pixels])
    
    assert len(pdist) == len(int_pixels)
    
    min_dist = 10000000
    value = None
    for i in range(len(pdist)):
        if min_dist > pdist[i]:
            min_dist = pdist[i]
            value = int_pixels[i]
            
    # int_pixels = np.array(int_pixels)
    # inds = pdist.argsort()
    #trash, value = min(zip(pdist, int_pixels), key=lambda x: x[0])
    return value #int_pixels[inds]

```

```python
pylab.rcParams['figure.figsize'] = (12, 12)
from timeit import default_timer as timer

img = mpimg.imread('data/1_4.bmp')

kernel_size = 5
plt.suptitle('kernel size = %d' % kernel_size, size=14)

plt.subplot(2, 2, 1)
start = timer()
imgplot = imshow(img)
end = timer()
elapsed = end - start
plt.title('Original Image')
print('finished showing original image %d' % elapsed)

plt.subplot(2, 2, 2)
start = timer()
filtered_img = kernel_filter(img, normal_ordering, kernel_size)
end = timer()
elapsed = end - start
print('finished normal ordering, %d' % elapsed)
imgplot = imshow(filtered_img)
plt.title('Normal ordering')

plt.subplot(2, 2, 4)
start = timer()
filtered_img = kernel_filter(img, petunin_ordering, kernel_size)
end = timer()
elapsed = end - start
print('finished petunin ordering, %d' % elapsed)
imgplot = imshow(filtered_img)
plt.title('Petunin ordering')

plt.subplot(2, 2, 3)
start = timer()
filtered_img = kernel_filter(img, marginal_ordering, kernel_size)
end = timer()
elapsed = end - start
print('finished marginal ordering, %d' % elapsed)
imgplot = imshow(filtered_img)
plt.title('Marginal ordering')
```
