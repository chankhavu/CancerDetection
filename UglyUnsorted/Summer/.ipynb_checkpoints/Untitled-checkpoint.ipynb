```python
from matplotlib import image as mpimg
from matplotlib import pyplot as plt

%pylab inline
```

```python
def kernel_filter(img, ordering, size):
    new_img = numpy.ndarray(img.shape, dtype=uint8)    
    def apply_kernel(img, new_img, ordering, x, y, size):     
        assert size % 2 == 1
        top = max(y-(size//2), 0)
        bottom = min(y+(size//2), img.shape[0])
        left = max(x-(size//2), 0)
        right = min(x+(size//2), img.shape[1])        
        
        pixels = []        
        for i in range(top, bottom):
            for j in range(left, right):
                pixels += [img[i][j]]
        value = ordering(pixels)
                
        pixel_value = value #pixels[len(pixels)//2 + 1]
        new_img[y][x] = pixel_value

    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            apply_kernel(img, new_img, ordering, j, i, size)
    return new_img

def normal_ordering(pixels):
    ordered = sorted(pixels, key=lambda point: 0.33*point[0] + 0.33*point[1] + 0.34*point[2])
    return ordered[len(ordered)//2]
```

```python
import subprocess
pylab.rcParams['figure.figsize'] = (16, 16)

input_imgname = 'data/8_1.bmp'
kernel_radius = 1

plt.suptitle('Kernel Size = %d' % (kernel_radius*2 + 1), size=14)

plt.subplot(2, 2, 1)
img = mpimg.imread(input_imgname)
imgplot = imshow(img)
plt.title('Original Image')
print('finish original')

plt.subplot(2, 2, 2)
filtered_img = kernel_filter(img, normal_ordering, (kernel_radius*2 + 1))
imgplot = imshow(filtered_img)
plt.title('Median Ordering')
print('finish normal')

plt.subplot(2, 2, 3)
process = subprocess.Popen('./cpp/marg %s temp.png %d' % (input_imgname, kernel_radius), 
                           shell=True, stdout=subprocess.PIPE)
process.wait()
petimg = mpimg.imread('temp.png')
imgplot = imshow(petimg)
plt.title('Marginal Ordering')
print('finish marginal')

plt.subplot(2, 2, 4)
process = subprocess.Popen('./cpp/pet %s temp.png %d' % (input_imgname, kernel_radius), 
                           shell=True, stdout=subprocess.PIPE)
process.wait()
petimg = mpimg.imread('temp.png')
imgplot = imshow(petimg)
plt.title('Petunin Ordering')
print('finish petunin')
```
